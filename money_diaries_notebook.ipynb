Engagement (measured in the number of comments on each diary) has been on an overall decline since September 2023. The months with the lowest average engagement in any year are May, July, and August. Day of week doesn't significantly impact engagement, though it does decline a bit throughout the average week (Monday diaries get an average of 74 comments while Friday diaries get an average of 64 comments).

Let's use predictive modeling in scikit-learn to see which factors drive engagement the most, to see which diaries Refinery29 may want to prioritize when deciding which of their submissions to publish

### Let's try linear regression to start, with R29 Number of Comments being the target variable

```python
md.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import make_column_selector, ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, PolynomialFeatures
from sklearn.linear_model import LinearRegression

X = md.drop(["Date", "R29 Number of Comments"], axis = 1)
y = md["R29 Number of Comments"]

ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lr_pipeline = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
)
```

```python
# Fit linear regression pipeline on full dataset and interpret some of the important coefficients
lr_pipeline_fitted = lr_pipeline.fit(X, y)
lr_coefs = lr_pipeline_fitted.named_steps["linear_regression"].coef_

feature_names = ct.get_feature_names_out()
pd.set_option("display.max_rows", None)
pd.set_option("display.float_format", "{:.6f}".format) # makes sure output is not in scientific notation
pd.Series(abs(lr_coefs), index=feature_names).sort_values(ascending=False) # I'm viewing the coefficients by absolute numbers first,
                                                                            # then once I narrow down the most important features, I will
                                                                            # set them back to positive and negative
```

```python
# Use cross-validation to estimate R-squared
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lr_pipeline, X, y, cv=5, scoring="r2")
scores.mean()
```

```python
# the top 5 variables seem to be Industry, City/Region, State, R29 Angry Reactions, and Income. Let's try a model with just those
X = md[["Industry", "City/Region", "State", "R29 Angry Reactions", "Income"]]
y = md["R29 Number of Comments"]

ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lr_pipeline = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
)

scores = cross_val_score(lr_pipeline, X, y, cv=5, scoring="r2")
scores.mean()
```

```python
# Much better than the one with all the variables! Now let's try the top 10, which include Joint, R29 Impressed Reactions, Age, Day of Week &
                                                                                                                          # Individual Income
X = md[["Industry", "City/Region", "State", "R29 Angry Reactions", "Income", "Joint", "R29 Impressed Reactions", "Age", "Day of Week",
        "Individual Income"]]
y = md["R29 Number of Comments"]

ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lr_pipeline = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
)

scores = cross_val_score(lr_pipeline, X, y, cv=5, scoring="r2")
scores.mean()
```

```python
# Even better! Now let's do the top 15
X = md[["Industry", "City/Region", "State", "R29 Angry Reactions", "Income", "Joint", "R29 Impressed Reactions", "Age", "Day of Week",
        "Individual Income", "Lives Alone", "Debt", "Housing Costs per Month", "R29 Surprised Reactions", "Total Spent in Week"]]
y = md["R29 Number of Comments"]

ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lr_pipeline = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
)

scores = cross_val_score(lr_pipeline, X, y, cv=5, scoring="r2")
scores.mean()
```

That one's almost as bad as the one with all the variables! I experimented by removing each of the last five one by one to see which ones were causing the score to go down.

The best feature set I could come up with was the top 10 plus Debt, which led to roughly an R-squared of -0.3544.

```python
X = md[["Industry", "City/Region", "State", "R29 Angry Reactions", "Income", "Joint", "R29 Impressed Reactions", "Age", "Day of Week",
        "Individual Income", "Debt"]]
y = md["R29 Number of Comments"]

ct = ColumnTransformer(
  [
    ("dummify",
    OneHotEncoder(sparse_output = False, handle_unknown="ignore"),
    make_column_selector(dtype_include=object)),
    ("standardize",
    StandardScaler(),
    make_column_selector(dtype_include=np.number))
  ],
  remainder = "passthrough"
)

lr_pipeline = Pipeline(
  [("preprocessing", ct),
  ("linear_regression", LinearRegression())]
)

scores = cross_val_score(lr_pipeline, X, y, cv=5, scoring="r2")
scores.mean()
```

### Let's try this same feature set but with Ridge regression

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge_pipeline = Pipeline(
  [("preprocessing", ct), # the same preprocessing ColumnTransformer as earlier works
  ("ridge", Ridge())]
)

# Use cross-validation to tune λ
param_grid = {
    "ridge__alpha": [0.001, 0.01, 0.1, 1, 10]
}

gscv = GridSearchCV(ridge_pipeline, param_grid, cv = 5, scoring="r2")
gscv_fitted = gscv.fit(X, y)
gscv_fitted.cv_results_["mean_test_score"]
```

A Ridge regression model with the best feature set and λ=10 would have an R-squared of 0.0314, a significant improvement from the linear regression model's -0.3544!

### Now let's try LASSO

```python
from sklearn.linear_model import Lasso
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning) # ConvergenceWarnings are creating noisy output


lasso_pipeline = Pipeline(
  [("preprocessing", ct),
  ("lasso", Lasso())]
)

param_grid = {
    "lasso__alpha": [0.001, 0.01, 0.1, 1, 10]
}

gscv = GridSearchCV(lasso_pipeline, param_grid, cv = 5, scoring="r2")
gscv_fitted = gscv.fit(X, y)
gscv_fitted.cv_results_["mean_test_score"]
```

A LASSO regression model with the best feature set and λ=1 would have an R-squared of 0.0598, the best so far!

### Finally, let's see if Elastic Net will give us the best of both worlds

```python
from sklearn.linear_model import ElasticNet

elastic_net_pipeline = Pipeline(
  [("preprocessing", ct),
  ("elastic_net", ElasticNet())])

param_grid = {
    "elastic_net__alpha": [0.001, 0.01, 0.1, 1, 10],
    "elastic_net__l1_ratio": [0.1, 0.3, 0.5, 0.7, 0.9, 1.0]
}

gscv = GridSearchCV(elastic_net_pipeline, param_grid, cv=5, scoring="r2")
gscv_fitted = gscv.fit(X, y)
print("Best params:", gscv_fitted.best_params_)
print("Best score:", gscv_fitted.best_score_)
```

```python
# Fit elastic net pipeline with the best parameters
elastic_net_pipeline = Pipeline([("preprocessing", ct), ("elastic_net", ElasticNet(alpha=1, l1_ratio=0.9))])
elastic_net_pipeline_fitted = elastic_net_pipeline.fit(X, y)

# Get the coefficients of this model for interpretation
feature_names = ct.get_feature_names_out()
elastic_net_coefs = elastic_net_pipeline_fitted.named_steps["elastic_net"].coef_
pd.Series(elastic_net_coefs, index=feature_names).sort_values(ascending=False)
```
